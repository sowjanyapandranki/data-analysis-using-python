```python
d1={'sno':[101,102,103,104,110],
   'sname':['a','b','d','s','e'],
    'sage':[10,20,30,40,50]}
d1
df1=pd.DataFrame(d1)
df1
```

```python
d2={'sno':[101,102,103,104,110],
   'saddress':['ad','bg','dg','sr','eg'],
    'smarks':[100,200,300,400,500]}
d2
df2=pd.DataFrame(d2)
df2
```

```python
df_merge=pd.merge(df1,df2,on='sno')
df_merge
```

```python
df_merge=pd.merge(df1,df2,on='sno',how='left')
df_merge
```

```python
df_outerjoin=pd.merge(df1,df2,on='sno',how='outer')
df_outerjoin
```
