### loc and iloc

* loc means location of index

* iloc means index location

```python
d={'sno':[101,102,103,104,105],'sname':['a','b','c','d','e'],'smarks':[24,55,76,87,3]}
df=pd.DataFrame(d)
df
```

```python
d={'sno':[101,102,103,104,105],'sname':['a','b','c','d','e'],'smarks':[24,55,76,87,3]}
df=pd.DataFrame(d)
df=df.set_index("sno")
df
```

```python
df.iloc[0:3,1:]
```

```python
df.loc[0:]
```

```python
df.loc[103:]
```

```python
df.loc[103:105]
```

```python
df.loc[103:105,'sname':]
```

### condition based access data

```python
df[df['smarks']>80]
```
