```python
x=np.linspace(0,4,50)
plt.plot(x,x,'r',label="first")
plt.plot(x,x**3,'b',label="second")
plt.legend(loc=(1.2,0.7))
plt.show()
```

```python
help(plt.legend())
```

```python
x=np.linspace(0,4,50)
plt.plot(x,x,'r',label="line")
plt.plot(x,x**2.2,'b*--',label='dashed line')
plt.plot(x,x**2.4,'r^',label='line caps')
plt.legend()
plt.show()
```

```python
x=np.linspace(0,10,50)
plt.plot(x,np.sin(x))
plt.plot(x,np.cos(x))
plt.title("sin vs cos")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
x=np.array([10,20,30,40])
y=np.array([50,55,32,67])
plt.scatter(x,y,color='red',marker='^')
plt.title("scatter plot")
plt.xlabel("x values")
plt.ylabel("y values")
for a,b in zip(x,y):
    plt.text(a,b,(str(a),str(b)))
plt.plot()
```

```python
x=np.array([10,20,30,40])
y=np.array([50,55,32,67])
plt.subplot(2,2,1)
plt.plot(x,y)
plt.subplot(2,2,2)
plt.plot(x,y,'r')
plt.subplot(2,2,3)
plt.plot(x,y,'m')
plt.subplot(2,2,4)
plt.plot(x,y,'b')
plt.show()
```

### Bar plot

```python
x=np.array(['chiru','nag','bala','venky'])
y=np.array([90,99,89,25])
plt.bar(x,y)
for a,b in zip(x,y):
    plt.text(a,b,(str(a),str(b)))
plt.show()
```

### Pie plot

```python
lang=['kova','conda','see','python','NET']
count=[90,100,67,98,40]
plt.pie(count,labels=lang,explode=[0.1,0,0,0,0],autopct='%1.2f%%')
plt.show()
```

```python
branch=['CSE','ECE','EEE','IT','MECH']
per=[95,86,67,78,40]
plt.pie(per,labels=branch,explode=[0.1,0,0,0,0],autopct='%1.2f%%')
plt.show()
```

### scikit-learn
scikit learn is a library contains the predefined datasets,machine learning algorithms,data preprocess thecniques

```python
from sklearn.datasets import load_iris
d=load_iris()
df=pd.DataFrame(d.data,columns=d.feature_names)
df['target']=d.target
df.head(2)
```

```python
l=[]
for i in df['target']:
    if(i==0):
        l.append('setosa')
    elif(i==1):
        l.append('versicolor')
    elif(i==2):
        l.append('verginica')
df['classes']=l
df.sample(10)
```

```python
sns.catplot(x='classes',y='sepal width (cm)',data=df)
plt.show()
```

```python
sns.catplot(x='classes',y='petal length (cm)',data=df)
plt.show()
```

```python
sns.catplot(x='classes',y='sepal width (cm)',data=df,kind='swarm')
plt.show()
```

```python
sns.catplot(x='classes',y='sepal width (cm)',data=df,kind='box')
plt.show()
```

### correlation:

it provides the relation between two features

* relation is 0 => their is no correlation between features


* relation is 1 => positive correlation

* relation is -11 => negative correlation

```python
cor=df.corr()
```

```python
cor
```

```python
sns.heatmap(cor,annot=True)
plt.show()
```
